### An√°lisis de los datos

#### Visitas a Showz

```python
# Obtener fecha, semana y mes de las visitas
visits_logs['session_date'] = pd.to_datetime(visits_logs['start_ts']).dt.date
visits_logs['session_week'] = pd.to_datetime(visits_logs['start_ts']).astype('datetime64[W]')
visits_logs['session_month'] = pd.to_datetime(visits_logs['start_ts']).astype('datetime64[M]')
```

```python
visits_logs.head()
```

```python
# Personas que usan Showz cada d√≠a
users_date = visits_logs.groupby('session_date')['uid'].nunique()
users_date
```

```python
# Personas que usan Showz cada semana
users_week = visits_logs.groupby('session_week')['uid'].nunique()
users_week
```

```python
# Personas que usan Showz cada mes
users_month = visits_logs.groupby('session_month')['uid'].nunique()
users_month
```

##### Duraci√≥n de las sesiones de los usuarios

```python
# Duraci√≥n de cada sesi√≥n (minutos)
visits_logs['start_ts'] = pd.to_datetime(visits_logs['start_ts'])
visits_logs['end_ts'] = pd.to_datetime(visits_logs['end_ts'])
visits_logs['session_duration_min'] = (visits_logs['end_ts'] - visits_logs['start_ts']).dt.seconds/ 60
visits_logs.head()
```

```python
# Duraci√≥n de las sesiones de usuarios de Showz
plt.figure(figsize=(12,5))
visits_logs['session_duration_min'].hist(bins=300)
plt.title('Duraci√≥n de las sesiones de usuarios de Showz')
plt.xlabel('Duraci√≥n de la sesi√≥n (minutos)')
plt.ylabel('Cantidad de sesiones')
plt.xlim(0,120)
```

```python
sns.boxplot(visits_logs['session_duration_min'])
plt.xlabel('Duraci√≥n de la sesi√≥n (minutos)')
plt.title('Duraci√≥n de las sesiones de Showz')
```

```python
print('Duraci√≥n promedio de la sesi√≥n (minutos)', round(visits_logs['session_duration_min'].mean() ,2))
print('Mediana de la duraci√≥n de la sesi√≥n', visits_logs['session_duration_min'].median())
print('Moda de la duraci√≥n de la sesi√≥n', visits_logs['session_duration_min'].mode()[0])
```

La gran mayor√≠a de los usuarios de Showz no mantienen una sesi√≥n de m√°s de 5 minutos aprox. De lo cual se podr√≠a deducir que estos usuarios no navegan mayormente por la p√°gina web; pueden ingresar al sitio unicamente a comprar las entradas que buscan.

##### Cohortes

```python
# Recuperar fecha de la primera visita para cada usuario
first_visits = visits_logs.groupby('uid')['session_date'].min().reset_index()
first_visits.columns= ['uid', 'first_session']
first_visits
```

```python
# Recuperar mes de la primera visita para cada usuario
first_visits['first_session_month'] = pd.to_datetime(first_visits['first_session']).astype('datetime64[M]')
```

```python
first_visits.head(3)
```

```python
# Agregar datos de primeras visitas a data de visitas de Showz
visits = pd.merge(visits_logs, first_visits, on='uid')
visits.sample(4)
```

```python
# Tiempo de vida de la sesion de cada usuario
visits['session_date'] = pd.to_datetime(visits['session_date'])
visits['first_session'] = pd.to_datetime(visits['first_session'])

visits['lifetime'] = (visits['session_date'] - visits['first_session']) / (np.timedelta64(1, 'M'))
visits['lifetime'] = visits['lifetime'].round().astype(int)
visits.head(3)
```

##### LTV

```python
# Cu√°nto dinero traen los clientes (LTV)
# Cantidad de clientes por mes 
cohort_clients = order_logs.groupby(['first_order_month', 'order_month']).agg({'uid':'nunique', 'revenue':'sum'}).reset_index()
cohort_clients.rename(columns={'uid':'n_clients', 'revenue':'total_revenue'}, inplace=True)
```

```python
# Calcular edad de la cohorte
cohort_clients['lifetime'] = (cohort_clients['order_month'] - cohort_clients['first_order_month']) / np.timedelta64(1, 'M')
cohort_clients['lifetime'] = cohort_clients['lifetime'].round().astype(int)
```

```python
# Calcular LTV
cohort_clients['ltv'] = cohort_clients['total_revenue'] / cohort_clients['n_clients']
```

```python
cohort_clients
```

```python
# Tabla din√°mica LTV
ltv = cohort_clients.pivot_table(
                            index='first_order_month',
                            columns='lifetime',
                            values='ltv',
                            aggfunc='sum'
)
ltv
```

Diciembre fue el mes con mayor cantidad de pedidos, pero al parecer los clientes de quienes se percibi√≥ un mayor ingreso fueron aquellos de cohortes previas a Dciiembre, es decir; aquellos que ya eran clientes; principalmente la cohorte de Septiembre 2017 que tuvo un ltv de 138 en Diciembre

##### ROMI

```python
# Qu√© tan rentables eran las inversiones?
# Cohortes mensuales:
cohort_clients
```

```python
cohort_clients
```

```python
# Calcular CAC
cohort_clients['cac'] = cohort_clients['costs'] / cohort_clients['n_clients']
cohort_clients.head()
```

```python
# Calcular ROMI
cohort_clients['romi'] = cohort_clients['ltv'] / cohort_clients['cac']
cohort_clients.head()
```

```python
# Calcular el ROMI acumulado
romi = cohort_clients.pivot_table(
    index='first_order_month',
    columns='lifetime',
    values='romi',
    aggfunc='mean'
)
romi_acum = romi.cumsum(axis=1).round(2)
romi_acum
```

S√≥lo 2 cohortes han recuperado la inversi√≥n; 
la primera cohorte, del 2017-06, comienza a recuperar la inversi√≥n al 7mo mes del ciclo de vida.  
La cohorte del 2017-09 comienza a recuperar la inversi√≥n al 4to mes del ciclo de vida.  
Sin embargo, las dem√°s cohortes, si bien van incrementando el retorno de lo invertido, no logran recuperar en el tiempo que llevan.

```python
# Cohorte promedio: promediar el ROMI de todas las cohortes para cada mes del ciclo de vida
romi.cumsum(axis=1).mean(axis=0)
```

En promedio, las cohortes comienzan a compensar lo invertido reci√©n al 9no - 10mo mes del ciclo de vida
